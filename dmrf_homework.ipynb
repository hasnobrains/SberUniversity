```python
import requests
```

###  Список айдишников новостроек

```python
offset = 0       # отступ в запросе
limit = 1000     # литмит на выдачу
objects_ids_total = [] # список айдишников всех новостроек
```

```python
while True:
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset}&limit={limit}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
    res = requests.get(url)
    data = res.json()['data']
    data_length = len(data['list'])
    if data_length > 0:
        objects_list = data['list']
        objects_ids = [x.get('objId') for x in objects_list]
        objects_ids_total += objects_ids
        print(f'objects total = {len(objects_ids_total)}')
        offset += limit
    else:
        break # если вернулся пустой список
```

```python
len(objects_ids_total)
```

Итого: 10202 новостройки

### Данные по каждому объекту

```python
table = [] # результирующая таблица
```

```python
for id in objects_ids_total:
    cell = {}
    print(f'Collecting object {id} info ...')
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{id}'
    res = requests.get(url)
    data = res.json()['data']
    try:
        cell['id'] = data.get('id')
        cell['region'] = data.get('region')
        cell['address'] = data.get('address')
        cell['name'] = data.get('nameObj')
        cell['floors'] = data.get('floorMin')
        cell['flats'] = data.get('objElemLivingCnt')
        cell['parking'] = data.get('objElemParkingCnt')
        cell['ready_date'] = data.get('objReady100PercDt')
        cell['material'] = data.get('wallMaterialShortDesc')
        cell['class'] = data.get('objLkClassDesc')
        cell['url'] = url
    except:
        break
        
    table.append(cell)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(table) # создаем датафрейм из таблицы
df.set_index('id', inplace=True)
df
```

Сохраняем в excel

```python
df.to_excel("новостройки.xls")  
```

Сохраняем в pickle

```python
df.to_pickle("новостройки.pkl")
```

Сохраняем в БД

```python
import sqlite3

conn = sqlite3.connect('buildings')

df.to_sql('new_buildings', conn, if_exists='replace')

conn.commit()
conn.close()
```

## Cкачиваем фотографии всех строящихся объектов с сайта "наш.дом.рф"

```python
from bs4 import BeautifulSoup
import json
import os.path
```

```python
for id in objects_ids_total:
    print(f'Collecting images for object {id} ...')
    !mkdir -p  {id}
    url = f'https://наш.дом.рф/сервисы/каталог-новостроек/объект/{id}'
    res  = requests.get(url).text 
    soup = BeautifulSoup(res, 'html5lib')

    data = json.loads(soup.find('script', type='application/json').text)

    # еще один способ получения информации об объекте
    # data['props']['initialState']['kn']['objectCard']['mainInfo']['data']

    objects_list = data['props']['initialState']['kn']['objectCard']['constructionProgress']['shortInfo']

    img_counter = 1 # если будут повторяющиеся имена у изображений, то сохранять новое изображение со счетчиком
    
    for obj in objects_list:
        image = requests.get(obj['src'])
        filename = obj['filename']
        path = './' + str(id) + '/'

        if os.path.isfile(path + filename):
            filename = filename[:-4] + '(' + str(img_counter) + ').jpg'
            img_counter += 1
        else:
            img_counter = 1
            
        print(f'   Saving {filename}')
        
        file = open(path + filename, 'wb')
        file.write(image.content)
        file.close()

    break # для проверки загрузки изображений одного объекта. Можно удалить и скачаются фотографии всех объектов
```
