1. Выведите количество пустых значений по каждой из колонок (название колонки, кол-во пустых значений)

```python
pd.read_sql(
    """
    SELECT sum(case when CRIM is null then 1 else 0 end) as CRIM,
           sum(case when ZN is null then 1 else 0 end) as ZN,
           sum(case when INDUS is null then 1 else 0 end) as INDUS,
           sum(case when CHAS is null then 1 else 0 end) as CHAS,
           sum(case when NOX is null then 1 else 0 end) as NOX,
           sum(case when RM is null then 1 else 0 end) as RM,
           sum(case when AGE is null then 1 else 0 end) as AGE,
           sum(case when DIS is null then 1 else 0 end) as DIS,
           sum(case when RAD is null then 1 else 0 end) as RAD,
           sum(case when TAX is null then 1 else 0 end) as TAX,
           sum(case when PTRATIO is null then 1 else 0 end) as PTRATIO,
           sum(case when B is null then 1 else 0 end) as B,
           sum(case when LSTAT is null then 1 else 0 end) as LSTAT,
           sum(case when MEDV is null then 1 else 0 end) as MEDV
    FROM boston
    """,
    con,
)
```

```python
pd.read_sql(
    """
    select Result
    from (
    with a as
    (SELECT min(CRIM) as min_CRIM,
           meadian_CRIM,
           min(ZN) as min_ZN,
           meadian_ZN,
           min(INDUS) as min_INDUS,
           meadian_INDUS,
           min(CHAS) as min_CHAS,
           meadian_CHAS,
           min(NOX) as min_NOX,
           meadian_NOX,
           min(RM) as min_RM,
           meadian_RM,
           min(AGE) as min_AGE,
           meadian_AGE,
           min(DIS) as min_DIS,
           meadian_DIS,
           min(RAD) as min_RAD,
           meadian_RAD,
           min(TAX) as min_TAX,
           meadian_TAX,
           min(PTRATIO) as min_PTRATIO,
           meadian_PTRATIO,
           min(B) as min_B,
           meadian_B,
           min(LSTAT) as min_LSTAT,
           meadian_LSTAT,
           min(MEDV) as min_MEDV,
           meadian_MEDV
    FROM boston,
    (SELECT CRIM as meadian_CRIM FROM boston order by CRIM LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT ZN as meadian_ZN FROM boston order by ZN LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT INDUS as meadian_INDUS FROM boston order by INDUS LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT CHAS as meadian_CHAS FROM boston order by CHAS LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT NOX as meadian_NOX FROM boston order by NOX LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT RM as meadian_RM FROM boston order by RM LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT AGE as meadian_AGE FROM boston order by AGE LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT DIS as meadian_DIS FROM boston order by DIS LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT RAD as meadian_RAD FROM boston order by RAD LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT TAX as meadian_TAX FROM boston order by TAX LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT PTRATIO as meadian_PTRATIO FROM boston order by PTRATIO LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT B as meadian_B FROM boston order by B LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT LSTAT as meadian_LSTAT FROM boston order by LSTAT LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2),
    (SELECT MEDV as meadian_MEDV FROM boston order by MEDV LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston) / 2))
    
    SELECT case when min_CRIM = meadian_CRIM then 'CRIM' else null end as Result FROM a
    union all
    SELECT case when min_ZN = meadian_ZN then 'ZN' else null end as Result FROM a
    union all
    SELECT case when min_INDUS = meadian_INDUS then 'INDUS' else null end as Result FROM a
    union all
    SELECT case when min_CHAS = meadian_CHAS then 'CHAS' else null end as Result FROM a
    union all
    SELECT case when min_NOX = meadian_NOX then 'NOX' else null end as Result FROM a
    union all
    SELECT case when min_RM = meadian_RM then 'RM' else null end as Result FROM a
    union all
    SELECT case when min_AGE = meadian_AGE then 'AGE' else null end as Result FROM a
    union all
    SELECT case when min_DIS = meadian_DIS then 'DIS' else null end as Result FROM a
    union all
    SELECT case when min_RAD = meadian_RAD then 'RAD' else null end as Result FROM a
    union all
    SELECT case when min_TAX = meadian_TAX then 'TAX' else null end as Result FROM a
    union all
    SELECT case when min_PTRATIO = meadian_PTRATIO then 'PTRATIO' else null end as Result FROM a
    union all
    SELECT case when min_B = meadian_B then 'B' else null end as Result FROM a
    union all
    SELECT case when min_LSTAT = meadian_LSTAT then 'LSTAT' else null end as Result FROM a
    union all
    SELECT case when min_MEDV = meadian_MEDV then 'MEDV' else null end as Result FROM a)
    where Result is not null
    """,
    con,
)
```

4. Выведите разницу между среднем количеством комнат(RM) в домах с самой дорогой стоимостью(MEDV) и 25 самыми дешевыми домами. 
Аналогично по 50, 100, 200, 300 самыми дешевыми домами. (кол-во домов(25,50,100,200,300), среднее кол-во комнат в них, среднее кол-во комнат в самых дорогих, разница).
Напишите влияет ли кол-во комнат на стоимость и как сильно.

```python
pd.read_sql(
    """
    SELECT
        count(RM) AS house_count,
        avg(RM) AS avg_rm,
        avg_rm_top,
        avg_rm_top- avg(RM) AS diff
    FROM
        (SELECT RM FROM boston ORDER BY medv LIMIT 25),
        (SELECT avg(rm) FILTER(WHERE medv = max(medv)) OVER() AS avg_rm_top FROM boston)
    union all
    SELECT
        count(RM) AS house_count,
        avg(RM) AS avg_rm,
        avg_rm_top,
        avg_rm_top- avg(RM) AS diff
    FROM
        (SELECT RM FROM boston ORDER BY medv LIMIT 50),
        (SELECT avg(rm) FILTER(WHERE medv = max(medv)) OVER() AS avg_rm_top FROM boston)
    union all
    SELECT
        count(RM) AS house_count,
        avg(RM) AS avg_rm,
        avg_rm_top,
        avg_rm_top- avg(RM) AS diff
    FROM
        (SELECT RM FROM boston ORDER BY medv LIMIT 100),
        (SELECT avg(rm) FILTER(WHERE medv = max(medv)) OVER() AS avg_rm_top FROM boston)
    union all
    SELECT
        count(RM) AS house_count,
        avg(RM) AS avg_rm,
        avg_rm_top,
        avg_rm_top- avg(RM) AS diff
    FROM
        (SELECT RM FROM boston ORDER BY medv LIMIT 200),
        (SELECT avg(rm) FILTER(WHERE medv = max(medv)) OVER() AS avg_rm_top FROM boston)
    union all
    SELECT
        count(RM) AS house_count,
        avg(RM) AS avg_rm,
        avg_rm_top,
        avg_rm_top- avg(RM) AS diff
    FROM
        (SELECT RM FROM boston ORDER BY medv LIMIT 300),
        (SELECT avg(rm) FILTER(WHERE medv = max(medv)) OVER() AS avg_rm_top FROM boston)
    """,
    con,
)
```

5. Выведите ранги значений колонки LSTAT(процент населения с более низким статусом) в домах с самой дорогой стоимостью (значение LSTAT, стоимость, ранг). Напишите какой вывод можно сделать по этим данным.

```python
pd.read_sql(
    """
    SELECT lstat, medv, dense_rank() OVER (PARTITION BY lstat ORDER BY medv DESC) rank_by_lstat
    FROM boston
    """,
    con,
)
```

```python
pd.read_sql(
    """
    select Result
    from (
    SELECT 
    case when avg(CRIM) filter(where chas = 1.0) > avg(CRIM) filter(where chas = 0.0) then 'CRIM' else null end as Result
    FROM boston
    union all
    SELECT
    case when avg(ZN) filter(where chas = 1.0) > avg(ZN) filter(where chas = 0.0) then 'ZN' else null end as Result
    FROM boston
    union all 
    SELECT 
    case when avg(INDUS) filter(where chas = 1.0) > avg(INDUS) filter(where chas = 0.0) then 'INDUS' else null end as Result
    FROM boston
    union all
    SELECT 
    case when avg(CHAS) filter(where chas = 1.0) > avg(CHAS) filter(where chas = 0.0) then 'CHAS' end as Result
    FROM boston
    union all
    SELECT 
    case when avg(NOX) filter(where chas = 1.0) > avg(NOX) filter(where chas = 0.0) then 'NOX' end as Result
    FROM boston
    union all
    SELECT 
    case when avg(RM) filter(where chas = 1.0) > avg(RM) filter(where chas = 0.0) then 'RM' end as Result
    FROM boston
    union all    
    SELECT 
    case when avg(AGE) filter(where chas = 1.0) > avg(AGE) filter(where chas = 0.0) then 'AGE' end as Result
    FROM boston
    union all     
    SELECT 
    case when avg(DIS) filter(where chas = 1.0) > avg(DIS) filter(where chas = 0.0) then 'DIS' end as Result
    FROM boston
    union all      
    SELECT 
    case when avg(RAD) filter(where chas = 1.0) > avg(RAD) filter(where chas = 0.0) then 'RAD' end as Result
    FROM boston
    union all   
    SELECT 
    case when avg(TAX) filter(where chas = 1.0) > avg(TAX) filter(where chas = 0.0) then 'TAX' end as Result
    FROM boston
    union all   
    SELECT 
    case when avg(PTRATIO) filter(where chas = 1.0) > avg(PTRATIO) filter(where chas = 0.0) then 'PTRATIO' end as Result
    FROM boston
    union all 
    SELECT 
    case when avg(B) filter(where chas = 1.0) > avg(B) filter(where chas = 0.0) then 'B' end as Result
    FROM boston
    union all 
    SELECT 
    case when avg(LSTAT) filter(where chas = 1.0) > avg(LSTAT) filter(where chas = 0.0) then 'LSTAT' end as Result
    FROM boston
    union all
    SELECT
    case when avg(MEDV) filter(where chas = 1.0) > avg(MEDV) filter(where chas = 0.0) then 'MEDV' else null end as Result
    FROM boston)
    where Result is not null
    """,
    con,
)
```
